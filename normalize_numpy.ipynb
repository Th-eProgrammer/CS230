```python
import sys
import os
import random
import os
import numpy as np

from shutil import copyfile
from PIL import Image
from tqdm import tqdm
```

```python
def normalize_255(a):
    a = a * (255.0 / a.max()) if a.max() != 0 else a
    a = a.astype(int)
    return a

os.chdir("/home/data/FETAL/val")
files = os.listdir(os.getcwd())
successes = 0
fails = 0
for file in files:
    try:
        a = np.load(file)
        print("a success")
        successes += 1
    except:
        print("file " + file + " could not be loaded")
        fails +=1
        continue
    new_a = normalize_255(a)
    os.system("rm " + file)
    np.save(file, new_a)

print(str(successes) + "successes " +str(fails) + " fails")
```
